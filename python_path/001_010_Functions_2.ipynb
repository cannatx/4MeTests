## LAMBDAS
the keyword `lambda` is used to define an anonymous function

`lambda <input_arguments>: <expression>`

Se pueden definir e invocar con argumentos igual que los de cualquier otra función, la unica diferencia es que las funciones lambda solo aceptan expresiones

Es muy util con la función `filter`

```python
cube_of = lambda x: x * x * x
cube_of(3)
```

```python
multiplier = lambda x,y: x*y
multiplier(2,4)
```

```python
func_dict = {
    "cube": lambda x: x * x * x,
    "squ": lambda x: x * x,
    "add": lambda x,y: x + y,
}

print(func_dict["cube"](5))
print(func_dict.get("add")(3,7))
```

```python
(lambda x,y: (x *x *x) - y)(3,2)
```

```python
(lambda *numbers: sum(numbers))(2,4,5,655,3,344,56,2)
```

```python
(lambda **num_dict: sum(num_dict.values()))(a=23, b=45, c=76, d=3)
```

```python
num_list = [2,453,6,72,23,3,5,76,324,3,2,3,6,2542,5,4]

num_list_result = [()(y) for y in num_list ]

num_list_result
```

```python
filter(lambda x: x >= 20, num_list)
```

```python
list(filter(lambda x: x >= 20, num_list))
```

```python
pares = [x for x in filter(lambda i: i % 2 == 0, num_list)]
impares = list(filter(lambda i: i % 2 != 0, num_list))
print(pares)
print(impares)
```
